```python
import numpy
import pandas as pd
import os
```

```python
# access data/metadata from sensors 
coords_df = pd.read_csv('/Users/emilyzhao/li-sound/data/used/coords.csv', low_memory=False)
nutrients_df = pd.read_csv('/Users/emilyzhao/li-sound/data/used/quality-nutrients.csv', low_memory=False)
```

```python
# add in coordinates

nutrients_df['LAT'] = ''
nutrients_df['LON'] = ''

for i in nutrients_df['STATION_ID']:
    for x in coords_df['STATION_ID']:
        if i == x:
            nutrients_df.loc[nutrients_df['STATION_ID'] == i, 'LAT'] = coords_df.loc[coords_df['STATION_ID'] == i, 'LATITUDE DD'].values[0]
            nutrients_df.loc[nutrients_df['STATION_ID'] == i, 'LON'] = coords_df.loc[coords_df['STATION_ID'] == i, 'LONGITUDE DD'].values[0]
```

```python
# return by surface/botttom

def depth_data(df, depth):

    if depth == 'surface':
        filtered = df[df['CLASS'] == 'Surface']
    elif depth == 'bottom':
        filtered = df[df['CLASS'] == 'Bottom']

    final = pd.DataFrame()
    #final['DATE'] = pd.to_datetime(filtered['DATE'])
    final['DATE'] = filtered['DATE']
    final['STATION_ID'] = filtered['STATION_ID']
    final['CLASS'] = filtered['CLASS']
    final['DEPTH_M'] = filtered['DEPTH_M']
    final['DISSOLVED_OXYGEN_MG_L'] = pd.to_numeric(filtered['DISSOLVED_OXYGEN_MG_L'], errors='coerce')
    final['TEMPERATURE_C'] = pd.to_numeric(filtered['TEMPERATURE_C'], errors='coerce')
    final['SALINITY'] = pd.to_numeric(filtered['SALINTY_PSU'], errors='coerce')
    final['LON'] = pd.to_numeric(filtered['LON'], errors='coerce')
    final['LAT'] = pd.to_numeric(filtered['LAT'], errors='coerce')
    
    return final
```

```python
surface = depth_data(nutrients_df, 'surface')
bottom = depth_data(nutrients_df, 'bottom')
```
